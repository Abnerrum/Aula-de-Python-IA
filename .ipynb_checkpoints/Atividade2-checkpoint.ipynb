1. Verificar número positivo
Peça ao usuário um número e verifique: Se o número for
maior que 0, mostre "Número positivo". (Use apenas
if).

```python
numero = int(input("Digite um número: "))

if numero > 0:
    print("Número positivo")

elif numero < 0:
    print("Número negativo")

else:
    print("Número é zero")
```

2. Verificar maior de idade
Peça a idade do usuário. Mostre "Maior de idade" se a
idade for maior ou igual a 18. (Use apenas if).

```python
nota=int(input("digite sua Nota:"))

if nota >=18:
    print("Maior Idade ")
    
elif nota < 18:
    print("Menor Idade")
```

3. Número maior que 10
Peça um número ao usuário. Se o número for maior que
10, mostre "Número maior que 10". (Use apenas if).


```python
numero = int(input("Digite um número: "))

if numero > 10:
    print("Número maior que 10")
```


4. Par ou Ímpar
Peça um número inteiro. Mostre "Número par" ou
"Número ímpar". (Use if e else).

```python
numero = int(input("Digite um número inteiro: "))

if numero % 2 == 0:
    print("Número par")
else:
    print("Número ímpar")
```
